```
#-*- coding:utf-8 -*-
with open("/content/voa.txt", "r") as f:
    data = f.read()
```

```
import re
data = re.split(r";|,|\?\s|;\s|,\s|.\s|\\n|\n|\|--", data) 
```

```
str_strip = {}
     

for i in range(len(data)):
 
  if data[i] in str_strip:
    
    str_strip[data[i]].append(i+1)
  else:
    str_strip[data[i]] = [i+1]
```
